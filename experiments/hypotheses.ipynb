```python
import os
import random
import textwrap
import pandas as pd
from IPython.display import display
```

```python
HYPOTHESIS = "regular"
ES = "d"
METHODS = ["classification", "llm"]
METHOD = "classification"

if METHOD == "classification":
    RESULTS_PATH = f"./classification/final/h_{HYPOTHESIS}_es_{ES}/outputs"
elif METHOD == "llm":
    RESULTS_PATH = f"./llm_zero_shot_prompting/h_{HYPOTHESIS}_es_{ES}/outputs"
else:
    raise ValueError(f"`METHOD` must be in {METHODS}")

RESULTS_GIV = list(set(x.split('.')[0].replace("_csv", "") for x in os.listdir(RESULTS_PATH)))
```

```python
DATA_PATH = os.path.join("../data/hypotheses/entry", f"h_{HYPOTHESIS}_es_{ES}.csv")
DATA = pd.read_csv(DATA_PATH, index_col=0).drop_duplicates()
GIV_PROPS = DATA.groupby("giv_prop").agg({"obs": "count"}).sort_values(by="obs", ascending=False).reset_index()
GIV_PROPS.giv_prop = GIV_PROPS.giv_prop.apply(lambda x: x.split('/')[-1])
GIV_PROPS = GIV_PROPS[GIV_PROPS.giv_prop.isin(RESULTS_GIV)]
GIV_PROPS
```

```python
VALS_GIV = list(GIV_PROPS.giv_prop.values)
CUT = len(VALS_GIV)//3

EX_GIV = {}
TYPES_GIV = ["high", "medium", "low"]
for index, key in enumerate(TYPES_GIV):
    random.seed(23)
    EX_GIV[key] = random.sample(VALS_GIV[index*CUT:(index+1)*CUT], 1)[0]
EX_GIV
```

```python
TYPE_GIV = "high"
if TYPE_GIV not in TYPES_GIV:
    raise ValueError(f"`TYPE_GIV` must be in {TYPES_GIV}")

suffix = "_csv" if METHOD == "llm" else ""
giv_id = EX_GIV[TYPE_GIV].split('/')[-1]

if METHOD == 'classification':
    df_output = pd.read_csv(os.path.join(RESULTS_PATH, f"{giv_id}{suffix}.csv"), index_col=0)
    cols = ['pred_num', 'pred_true', 'score_0.0', 'score_2.0', 'hypothesis']
    display(df_output[cols])

if METHOD == 'llm':
    with open(os.path.join(RESULTS_PATH, f"{giv_id}_readable.txt")) as openfile:
        hypotheses = [x.replace('\n', '') for x in openfile.readlines()]
    for h in hypotheses:
        print(f"{textwrap.fill(h, width=70)}\n=====")
```
